```python
import pandas as pd
import bs4
import selenium
import requests
```

```python
from selenium import webdriver
browser = webdriver.Chrome('chromedriver.exe')
browser.get('https://www.naukri.com/data-science-jobs')
```

```python
#for class use '.' and for id use '#' and '[]' for others
jobs = browser.find_elements_by_css_selector('div.row[type = "tuple"]') 
print(len(jobs))
```

```python
for job in jobs:
    
    title = job.find_element_by_css_selector('li.desig')
    title = title.get_attribute('title')  ## no need for inner html cuz title is not in between tags but its title itself
    
    loc_parent = job.find_element_by_css_selector('span.loc')
    loc_child = loc_parent.find_element_by_css_selector('span')
    
    try:
        loc = loc_child.find_element_by_css_selector('span.loc')
        loc = loc.get_attribute('innerHTML')
    except:
        loc = loc_child.get_attribute('innerHTML')
    
    org = job.find_element_by_css_selector('span.org')
    org = org.get_attribute('innerHTML')
    
    sal = job.find_element_by_css_selector('span.salary')
    sal = sal.get_attribute('innerHTML')
    
    exp = job.find_element_by_css_selector('span.exp')
    exp = exp.get_attribute('innerHTML')

    
    print(title, '|', org, '|', loc,'|', sal, '|', exp)
    print('---------')
```

```python
loc = browser.find_elements_by_css_selector('span.loc')
print(len(loc))
```

```python
df_naukri = pd.DataFrame(columns=['Title','Company','Experience','Salary','Date','Location','Skills','Description','URL'])

def get_info(job, selector, attribute):
    try:
        tag = job.find_element_by_css_selector(selector)
        tag = tag.get_attribute(attribute)
    except:
        tag = None
    return tag

for job in jobs:
    title = get_info(job, 'li.desig', 'title')
    org = get_info(job, 'span.org', 'innerHTML')
    exp = get_info(job, 'span.exp', 'innerHTML')
    sal = get_info(job, 'span.salary', 'innerHTML')
    date = get_info(job, 'span.date', 'innerHTML')
    skill = get_info(job, 'span.skill', 'innerHTML')
    description = get_info(job, 'span.desc', 'innerHTML')
    url = get_info(job,'a.content','href')
    
    loc_parent = job.find_element_by_css_selector('span.loc')
    location = get_info(loc_parent, 'span', 'innerHTML')
    
    curr_job = {
        'Title': title,
        'Company': org,
        'Experience': exp,
        'Salary': sal,
        'Date': date,
        'Location': location,
        'Skills': skill,
        'Description':description,
        'URL': url
    }
    
    df_naukri = df_naukri.append(curr_job, ignore_index=True)
df_naukri
    #print(title,'|', org, '|', exp,'|', sal, '|', date,'|', location)
    #print('-----------')
```

```python
df_naukri.to_csv('naukri_scrapping.csv', index = False)
```
