```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome('chromedriver.exe')
```

```python
myntra_url = "https://www.myntra.com/under-599-store"
browser.get(myntra_url)
```

```python
products = browser.find_elements_by_css_selector('li.product-base')
```

```python
len(products)
```

```python
for item in products:
    name = item.find_element_by_css_selector('div.product-brand')
    name = name.get_attribute('innerHTML') # to get text content between tags
    print(name)
```

```python
for item in products:
    name = item.find_element_by_css_selector('div.product-brand')
    name = name.get_attribute('innerHTML') # to get text content between tags
    desc = item.find_element_by_css_selector('h4.product-product')
    desc = desc.get_attribute('innerHTML')
    print(name, '|', desc)
```

```python
df_prod = pd.DataFrame(columns=['name','description','price','discount'])

for item in products:
    name = item.find_element_by_css_selector('div.product-brand')
    name = name.get_attribute('innerHTML') # to get text content between tags
    
    desc = item.find_element_by_css_selector('h4.product-product')
    desc = desc.get_attribute('innerHTML')
    
    discount = item.find_element_by_css_selector('span.product-discountPercentage')
    discount = discount.get_attribute('innerHTML')
    
    #print(name, '|', desc, '|', discount)
    
    price_parent = item.find_element_by_css_selector('div.product-price') ## price is diff so check parent name
    
    price_child = price_parent.find_element_by_css_selector('span')
    
    try:
        price = price_child.find_element_by_css_selector('span.product-discountedPrice')
        price = price.get_attribute('innerHTML')
    except:
        price = price_child.get_attribute('innerHTML')
    curr_row = {
        'name': name,
        'description': desc,
        'price':price,
        'discount': discount
    }
    df_prod = df_prod.append(curr_row, ignore_index=True)
df_prod
    
    #print(name, price)
    #print('-------')
```

```python
df_prod.to_csv('myntra_scrapping.csv', index = False)
```
